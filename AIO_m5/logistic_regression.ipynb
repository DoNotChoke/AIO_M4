```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1.4],[1.], [1.5], [3.], [3.8], [4.1]])
y = np.array([0.,0.,0.,1.,1.,1.])
```

```python
print(X)
print(y)
```

```python
b = np.ones((len(X),1))
X = np.concatenate((b,X), axis=1)
```

```python
print(X)
```

```python
w,b = -0.1, 0.1
theta = np.array([[b],[w]])
lr = 0.1
losses = []
```

```python
def predict(x, theta):
    z = x.T.dot(theta)
    y_hat = 1/(1+np.exp(-z))
    return y_hat
```

```python
def compute_loss(y, y_hat):
    return -y*np.log(y_hat) - (1-y)*np.log(1-y_hat)
```

```python
def gradient(x,y, y_hat):
    dtheta = x*(y_hat - y)
    return dtheta
```

```python
def update(theta, lr, dtheta):
    theta = theta - lr*dtheta
    return theta
```

```python
y.shape
```

```python
for _ in range(1000):
    for i in range(6):
        xi = X[i: i+1].reshape(len(theta),1)
        yi = y[i: i+1]
        y_hat = predict(xi, theta)
        
        loss = compute_loss(yi, y_hat)
        losses.append(loss[0])
        dtheta = gradient(xi, yi, y_hat)
        theta = update(theta, lr, dtheta)
```
